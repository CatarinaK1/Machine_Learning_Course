```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data from the excel file
# Obtained from kaggle https://www.kaggle.com/datasets/vipullrathod/fish-market/data
df_rawData = pd.read_csv('fish.csv')

# 'Species' would be our target.
```

```python
#Getting the unique values of fish Species.

SpeciesData = df_rawData["Species"].unique()

SpeciesData
```

```python
# Filtering the data to help with possible slicing 
bream_data = df_rawData[df_rawData['Species'] == 'Bream']
bream_data
```

```python
# Fetching the index location from start to end for roach

roach_data = df_rawData[df_rawData['Species'] == 'Roach']
roach_data
```

```python
# Fetching the index location from start to end for Whitefish

Whitefish_data = df_rawData[df_rawData['Species'] == 'Whitefish']
Whitefish_data
```

```python
# Fetching the index location from start to end for Parkki

Parkki_data = df_rawData[df_rawData['Species'] == 'Parkki']
Parkki_data
```

```python
#'Perch', 'Pike', 'Smelt'
# Fetching the index location from start to end for Perch

Perch_data = df_rawData[df_rawData['Species'] == 'Perch']
Perch_data
```

```python
# Fetching the index location from start to end for Pike

Pike_data = df_rawData[df_rawData['Species'] == 'Pike']
Pike_data
```

```python
# Fetching the index location from start to end for Pike

Smelt_data = df_rawData[df_rawData['Species'] == 'Smelt']
Smelt_data
```

## Fish Weight vs Fish Height comparison between all fish

```python
plt.xlabel('Fish Weight')
plt.ylabel('Fish Height')


plt.scatter(bream_data['Weight'], bream_data['Height'],color="green",marker='+')
plt.scatter(roach_data['Weight'], roach_data['Height'],color="blue",marker='.')
plt.scatter(Whitefish_data['Weight'], Whitefish_data['Height'],color="orange",marker='.')
plt.scatter(Parkki_data['Weight'], Parkki_data['Height'],color="pink",marker='o')
plt.scatter(Perch_data['Weight'], Perch_data['Height'],color="magenta",marker='D')
plt.scatter(Pike_data['Weight'], Pike_data['Height'],color="plum",marker='H')
plt.scatter(Smelt_data ['Weight'], Smelt_data['Height'],color="purple",marker='*')

```

## Fish length 1 vs Fish Width comparison between all fish

```python

plt.scatter(bream_data['Length1'], bream_data['Width'],color="green",marker='+')
plt.scatter(roach_data['Length1'], roach_data['Width'],color="blue",marker='.')
plt.scatter(Whitefish_data['Length1'], Whitefish_data['Width'],color="orange",marker='.')
plt.scatter(Parkki_data['Length1'], Parkki_data['Width'],color="pink",marker='o')
plt.scatter(Perch_data['Length1'], Perch_data['Width'],color="magenta",marker='D')
plt.scatter(Pike_data['Length1'], Pike_data['Width'],color="plum",marker='H')
plt.scatter(Smelt_data ['Length1'], Smelt_data['Width'],color="purple",marker='*')
```
